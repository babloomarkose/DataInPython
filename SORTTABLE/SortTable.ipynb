PART 2: VALIDATE DATA
Validation of data is done and data is cleaned in this part

      2.GPA is set to 2 decimal places & between values 1 and 4(values below 1 imputed with 1 & above with 4)

PART 3: SORT DATA and SORTING RULE
Inbuilt function Table.sort() from data science package's table module is used in sorting.
The table is sorted in ascending order of 'LastName', ascending is default argument to the sort() function. 
The colomn named 'LastName' is passed as argument as we need to sort the table based on that colomn. Sort function sorts the table in-place 
If the LastNames are same then the data is then sorted according to FirstName. To achieve this, as first step, sort according to First name so that sorting records with same last name will be sorted in ascending order of FirstName 

PART 4: WRITE DATA TO FILE
The sorted table instance is saven to a csv file locally. By default the current directory is choosen to save the file. If one needs to change the location of saved file, one need to update the variable 'outputfilepath' with one's choosen directory-path(without the '/' at the end)


```python
from datascience import Table # Including Table functions from datascience package into the running instance
import os# including os package into running instance

#LOAD DATA

#Loading data from CSV file URL address 
InputFilepath='https://raw.githubusercontent.com/babloomarkose/DataInPython/master/SortTable'
StudentInfo=Table().read_table(InputFilepath+'/StudentRecord.csv')

#Print the Student Table
print("The Unsorted Student Table:1st 20 rows ")
StudentInfo.show(20) # Show only the first 20 records

#DATA CLEANING is a necessary step before data manipulation

#1. Data Cleaning : Removing space from colomn label
StudentInfo=StudentInfo.relabeled(' FirstName', 'FirstName')


#2.  Data Cleaning : Set GPA to 2 decimal places and between values 1 and 4            
def ValidateGPA(x):
    if(x<1):
        return(1)
    elif(x>4):
        return(4)
    else:
        return(round(x,2))
StudentInfo['GPA'] =StudentInfo.apply(lambda x: ValidateGPA(x), 'GPA')

                        
#3. Data Cleaning : Set Credithours to whole number and between values 1 and 130
    #Limiting between values 1 and 130 as 1-130 for Bachelors and 1-64 for Masters, maximum being 130)           
def ValidateCredithours(x):
    if(x<1):
        return round(max(x, 1))  
    else:
        return round(min(x, 130))
StudentInfo['CreditHours'] =StudentInfo.apply(lambda x: ValidateCredithours(x), 'CreditHours')
                              
#4. Data Cleaning : Remove Duplicate records
pandasDF=StudentInfo.to_df()
pandasDF.drop_duplicates(keep='first',inplace=True)#keep='first' retains the first among n duplicates and inplace=True modify original record

#5. Data Cleaning : Impute Missing Values for colomns Mid_initail and Major
                   #'N.A' for Mid_initail and 'UNT' for Major. 
pandasDF['Mid_Initial']=pandasDF['Mid_Initial'].replace('nan','N.A')
pandasDF['Major']=pandasDF['Major'].replace('nan','UNT')


#6. Data Cleaning : Remove rows with missing values for any colomn other than colomns Mid_initail and Major
                   #For strings get the index of cell with 'nan' and drop the rows
                   #For nummeric values use built-in dropna()
indexNames = pandasDF[ pandasDF['LastName'] == 'nan' ].index
pandasDF.drop(indexNames , inplace=True)
indexNames = pandasDF[ pandasDF['Address'] == 'nan' ].index
pandasDF.drop(indexNames , inplace=True)
indexNames = pandasDF[ pandasDF['FirstName'] == 'nan' ].index
pandasDF.drop(indexNames , inplace=True)
pandasDF.dropna(subset=['GPA', 'Scholarship','CreditHours'], inplace=True)

#7.    #Data Cleaning : Set all field's Maximum width
pandasDF['LastName'] = pandasDF['LastName'].apply(lambda x: x[:10])
pandasDF['FirstName'] = pandasDF['FirstName'].apply(lambda x: x[:10])
pandasDF['Mid_Initial'] = pandasDF['Mid_Initial'].apply(lambda x: x[:1])
pandasDF['Address'] = pandasDF['Address'].apply(lambda x: x[:20])
pandasDF['Major'] = pandasDF['Major'].apply(lambda x: x[:15])
#Converting back to datascience table after data manipulation
StudentInfo1=Table.from_df(pandasDF)

#8. Data Cleaning : Convert Scholarship to TRUE/FALSE instead of 0/1
def ValidateScholarship(x):
    if(x==1):
        return ('TRUE')
    else:
        return ('FALSE')
StudentInfo1['Scholarship'] =StudentInfo1.apply(lambda x: ValidateScholarship(x), 'Scholarship')

#SORT
#Sort the table according to Last name, default is Ascending order
    #As first step sort according to First name so that sorting records with same last name will be sorted
    #in ascending order of FirstName. 
 
SortedTable=(StudentInfo1.sort('FirstName')).sort('LastName')
print("The Cleaned Student Table sorted according to Last Name")
SortedTable.show()

#Write output to file
outputfilepath = os.getcwd()
print(outputfilepath)
SortedTable.to_csv(outputfilepath+'/outputTable.csv')

#Displaying Sorted Student Data from output file
print("Reading from Output file The Cleaned and SORTED new Student FILE 20 rows")
StudentTable=Table.read_table(dirpath+'/outputTable.csv')
StudentTable.show(20)

```
