How to estimate a value of π using the Monte Carlo method - generate a large number of random points and see how many fall in the circle enclosed by the unit square.

One method to estimate the value of π (3.141592...) is by using a MonteCarlo method. We can have a circle of radius 0.5, enclosed by a 1 × 1 square.The area of the circle is π ∗ r2 = π/4, the area of the square is 1. If we divide
the area of the circle, by the area of the square we get π/4.We then generate a large number of uniformly distributed random points and plot them on the graph. These points can be in any position within the square i.e. between (0,0) and (1,1). If they fall within the circle, they are coloured red, otherwise they are coloured blue. We keep track of the total number of points,and the number of points that are inside the circle. If we divide the number of points within the circle, Ninner by the total number of points, Ntotal, we should get a value that is an approximation of the ratio of the areas we calculated above, π/4. In other words: π/4 = Ninner/Ntotal, and π = 4 × Ninner/Ntotal. When we only have a small number of points, the estimation is not very accurate, but when we have hundreds of thousands of points, we get much closer to the actual value - to within around 2 decimal places of accuracy.

Write a Python program to simulate the calculation of π using Monte Carlo method. Your program draws the circle and the square, and plot the random points in the graph as shown in Figure 1. You need simulate different numbers of points (at least include 10, 100, 1,000, 5,000, 10,000, 50,000, 100,000, 500,000,1,000,000, 5,000,000) and calculate the value π for each different number of points. Submit Python code, and simulation results including the graph plots and π values.

```python
from datascience import *
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

%matplotlib inline

#Function to determine if a point is outside or not of a circle with center 0.5,0.5 and radius 0.5
def IsPointInsideOrOnCircle(x,y):
    distance=np.sqrt(((0.5-x)**2)+((0.5-y)**2))
    if distance<=0.5:
        return True
    else:
        return False
    
#Function to create a Table of points, one column representing xCoordinate and other yCoordinate
def CreateTable(n):
    ArrayForXcordinate=np.random.uniform(0, 1,n)
    ArrayForYcordinate=np.random.uniform(0, 1,n)
    PointsTable = Table().with_columns('XCoordinate', ArrayForXcordinate,'YCoordinate', ArrayForYcordinate)
    return PointsTable

#Function to add a new coloumn to a table with value True/False based on 
#if the point represented by that row is inside/on or outside the circle 
def CreateColomnInsideorOnCircle(PointsTable):
    PointsTable.append_column('InsideorOnCircle',1)
    PointsTable['InsideorOnCircle']=PointsTable.apply(lambda x,y: 'True' if IsPointInsideOrOnCircle(x,y)==True else 'False','XCoordinate','YCoordinate')
    return PointsTable

#Function to plot the points and group then based on whether the point is inside or outside the circle    
def plotPoints(PointsTable):
    PointsTable.scatter('XCoordinate','YCoordinate',group='InsideorOnCircle')

#Function to calculate pi based on n points inside a unit square using a unit circle(Monte Carlo Method)    
def CalculatePi(n):
    #Create a Table with n rows and 2 coloumns for x,y coordinates of points each between 0 and 1
    CordinateTable=CreateTable(n)
    #Add a coloumn InsideorOnCircle, specifying if the point given by XCoordinate,YCoordinate is inside the unit circle
    CordinateTable=CreateColomnInsideorOnCircle(CordinateTable)
    #Plot the points represented by x and y coordinate in the Table
    plotPoints(CordinateTable)
    #Draw the circle with radius 0.5 and centered on (0.5,0.5)
    circle=plt.Circle((0.5, 0.5), 0.5, color='red',fill=False)
    fig = plt.gcf() # Get the current figure
    ax = fig.gca() #Get current axis
    ax.add_artist(circle)#Add the circle to current figure
    ax.set(title="Circle Inscribed in a Unit Square")
    plt.plot([0, 1,1,0,0], [0, 0,1,1,0])

    #Get the number of points inside the circle
    counter=Counter(CordinateTable['InsideorOnCircle'] )['True']
    pi=4*(counter/n)
    print('Value of Pi for n =',n,'is',pi)

#Get User input on the number of points to be considered
#two choice include either custom value or a predefined list of values
print("Do you Want to run for 10, 100, 1000, 5000, 10000, 50000, 100000, 500000,1000000, 5000000 number of points?")
print("It can take a couple of minutes")
print("Or you can check for a custom value")
print("Enter 1 for running with all the values in the list")
print("Enter 2 for entering a custom value")

Choice = int(input())
if Choice==1:
    NArray=[10, 100, 1000, 5000, 10000, 50000, 100000, 500000,1000000, 5000000]
    #Calculate pi for each value(giving number of points to be considered) in list
    for i in NArray:
        CalculatePi(i)
elif Choice==2:
    #Get the number of points to be considered to determine Pi
    print("Enter the number of points to be considered to determine Pi")
    n = int(input())
    CalculatePi(n)
else:
    print("Entered choice is invalid. Please Try Again")

```
